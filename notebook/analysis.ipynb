```python
# Ligands / drugs that interact on the same target, potential interaction
shared_targets = df.groupby(['Target', 'Target ID'])['Ligand'].unique().reset_index()
shared_targets['Ligand Count'] = shared_targets['Ligand'].apply(len)
shared_targets = shared_targets.sort_values(by='Ligand Count', ascending=False)
display(shared_targets)
```
